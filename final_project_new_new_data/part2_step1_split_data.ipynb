````python
import pandas
import sklearn.model_selection
````

# Instructions

- Read the data from CSV file and properly set the index
- Split the data into a train data set and a test data set using `sklearn.model_selection.train_test_split` with the following configurations:
```
    train_size   = 0.75
    shuffle      = True 
    random_state = 0
```
- Write the train data set to `./data/swiss_roll.train.csv`
- Write the test data set to `./data/swiss_roll.test.csv`
- **No data processing shall be performed in this step**


````python
data_train_test = pandas.read_csv('./data/high_salary.train.csv').set_index('id')
data_train_test
````

````python
data_train, data_test = sklearn.model_selection.train_test_split(
    data_train_test,
     train_size   = 0.75,
    shuffle      = True ,
    random_state = 0,
)
data_train.to_csv('./data/swiss_roll.train.csv')
data_test.to_csv('./data/swiss_roll.test.csv')
````
